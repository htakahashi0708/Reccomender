```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import math
```

```python
test_df = pd.read_parquet("data/test.parquet")
```

```python
train_df = pd.read_parquet("data/train.parquet")
```

```python
member_df = pd.read_parquet("data/members.parquet")
```

```python
song_df = pd.read_parquet("data/songs.parquet")
```

```python
song_df.dtypes
```

```python
n_song = song_df.shape[0]
n_song
```

```python
n_member = member_df.shape[0]
```

```python
member_df["msno_idx"] = np.arange(member_df.shape[0])
```

```python
msno_array = member_df["msno"].to_numpy()
```

```python
member_idx_dic = {}
for i in range(n_member):
    member_idx_dic[msno_array[i]] = i
```

```python
member_idx_dic
```

```python
def msno_convert(val, dic):
    try:
        val = dic[val]
    except KeyError:
        val = -1
        
    return val
```

```python
train_df["msno_idx"] = train_df["msno"].apply(lambda x: msno_convert(x,member_idx_dic ))
```

```python
test_df["msno_idx"] = test_df["msno"].apply(lambda x: msno_convert(x,member_idx_dic ))
```

```python
train_df.dtypes
```

```python
song_df
```

```python
song_df["artist_id"] = song_df["artist_name"].astype("category").cat.codes
```

```python
song_idx_dic = {}
song_id_array = song_df["song_id"].to_numpy()
for i in range(n_song):
    song_idx_dic[song_id_array[i]] = i
```

```python
train_df["song_idx"] = train_df["song_id"].apply(lambda x: msno_convert(x,song_idx_dic ))
```

```python
test_df["song_idx"] = test_df["song_id"].apply(lambda x: msno_convert(x,song_idx_dic ))
```

```python
song_df["song_idx"] = np.arange(n_song)
```

```python
train_df["song_idx"].max()
```

```python
song_all_set = set(pd.concat([train_df["song_idx"], test_df["song_idx"]]).to_numpy())
```

```python
member_all_set = set(pd.concat([train_df["msno_idx"]]).to_numpy())
```

```python
song_df["in_data"] = song_df["song_idx"].apply(lambda x: x in song_all_set)
```

```python
member_df["in_traindata"] = member_df["msno_idx"].apply(lambda x: x in member_all_set)
```

```python
filter1 =song_df["in_data"] 
```

```python
song_df = song_df.loc[filter1]
```

```python
n_song2 = song_df.shape[0]
```

```python
song_df["itemID"] = np.arange(n_song2)
```

```python
train_df = train_df.merge(song_df[["song_idx", "itemID"]], on = "song_idx", how = "left")
```

```python
test_df = test_df.merge(song_df[["song_idx", "itemID"]], on = "song_idx", how = "left")
```

```python
filter1 = train_df["itemID"].isna()
train_df.loc[filter1, "itemID"] = -1
```

```python
filter1 = test_df["itemID"].isna()
test_df.loc[filter1, "itemID"] = -1
```

```python
train_df["itemID"] = train_df["itemID"].astype(int)
```

```python
test_df["itemID"] = test_df["itemID"].astype(int)
```

```python
#member_df.drop(["in_data"], axis = 1, inplace = True)
```

```python
n_train = train_df.shape[0]
```

```python
n_test = test_df.shape[0]
```

```python
train_df["time"] = np.arange(n_train)
```

```python
test_df["time"] = np.arange(n_train, n_train + n_test)
```

```python
time_first_df = pd.concat([train_df[["song_idx", "time"]], test_df[["song_idx", "time"]]]).groupby("song_idx").min().reset_index().rename({"time":"time_first"}, axis = 1)
```

```python
train_df = train_df.merge(time_first_df, on = "song_idx", how = "left")
```

```python
test_df = test_df.merge(time_first_df, on = "song_idx", how = "left")
```

```python
train_df["time_old"] = train_df["time"] - train_df["time_first"]
```

```python
test_df["time_old"] = test_df["time"] - test_df["time_first"]
```
